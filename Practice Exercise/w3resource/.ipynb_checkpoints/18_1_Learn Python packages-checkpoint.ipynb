Write a Python program to extract all the URLs from the webpage python.org that are nested within \<li> tags from .

```python
import requests
from bs4 import BeautifulSoup
```

```python

url = 'https://www.python.org/'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'lxml')

urls = []
for h in soup.find_all('li'):
    a = h.find('a')
    urls.append(a.attrs['href'])
print(urls)
```

Write a  Python program to a list of all the h1, h2, h3 tags from the webpage  python.org.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.python.org/'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'lxml')

print("List of all the h1, h2, h3 :")
for heading in soup.find_all(["h1", "h2", "h3"]):
    print(heading.name + ' ' + heading.text.strip())
```

Write a Python program to test if a given page is found or not on the server.

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup("<b>w3resource.com</b>", "lxml")

print("Original Markup:")
print(soup.b)

tag = soup.new_tag("i")
tag.string = "Python"

print("\nNew Markup, after inserting the text:")
soup.b.string.insert_after(tag)
print(soup.b)
```

Write a  Python program to remove a tag or string from a given tree of html document and replace it with the given tag or string.

```python
from bs4 import BeautifulSoup

html_markup= '<a href="https://w3resource.com/">Python exercises<i>w3resource</i></a>'

soup = BeautifulSoup(html_markup, "lxml")
print("Original markup:")
a_tag = soup.a
print(a_tag)
new_tag = soup.new_tag("b")
new_tag.string = "PHP"
b_tag = a_tag.i.replace_with(new_tag)
print("New Markup:")
print(a_tag)
```

Write a  Python program to extract a tag or string from a given tree of html document.

```python
from bs4 import BeautifulSoup

html_content = '<a href="https://w3resource.com/">Python exercises<i>w3resource</i></a>'

soup = BeautifulSoup(html_content, "lxml")

print("Original Markup:")
print(soup.a)
i_tag = soup.i.extract()
print("\nExtract i tag from said html Markup:")
print(i_tag)
```

Write a  Python program to wrap an element in the specified tag and create the new wrapper.

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup("<p>Python exercises.</p>", "lxml")

print("Original Markup:")
print(soup.p.string.wrap(soup.new_tag("i")))
print("\nNew Markup:")
print(soup.p.wrap(soup.new_tag("div")))
```

Write a Python program to replace a given tag with whatever’s inside a given tag.

```python
from bs4 import BeautifulSoup

markup = '<a href="https://w3resource.com/">Python exercises.<i>w3resource.com</i></a>'

soup = BeautifulSoup(markup, "lxml")

a_tag = soup.a
print("Original markup:")
print(a_tag)

a_tag.i.unwrap()
print("\nAfter unwrapping:")
print(a_tag)
```

Write a Python program to get the number of datasets currently listed on data.gov.

```python
import requests
import bs4
```

```python
res = requests.get("https://data.gov/")
soup = bs4.BeautifulSoup(res.text,'html.parser')
data = soup.select('.text-color-red')
print(f"Datasets Available are: {data[0].getText()}")
```

Write a Python program to display the name of the most recently added dataset on data.gov.

```python
res = requests.get("https://catalog.data.gov/dataset?q=&sort=metadata_created+desc")
soup = bs4.BeautifulSoup(res.text,'html.parser')
```

```python
data1 = soup.select('.dataset-list.unstyled li')
```

```python
data2 = data1[0].select('a')
```

```python
print(f"Recently added Dataset is: {data2[0].getText()}")
print(f"Dataset Link: { 'https://catalog.data.gov/' + data2[0].attrs['href']}")
```

```python
for i in range(10):
    data = data1[i].select('a')
    print(f"Dataset is: {data[0].getText()}")
    print(f"Dataset Link: { 'https://catalog.data.gov/' + data[0].attrs['href']}\n")
```
